## 垃圾短信分类是指利用NLP技术，通过对短信文本内容进行建模，从而自动将短信分类为垃圾短信和正常短信。

## 通常包括以下步骤：

数据收集与预处理：收集大量的短信数据，并对数据进行清洗和预处理，包括去除重复数据、分词、去除停用词等。

特征提取：从短信文本中提取特征，这些特征可以包括词频、词性、句法结构等。

模型训练：使用机器学习或深度学习模型进行训练，常用的模型包括朴素贝叶斯、支持向量机、深度神经网络等。训练过程中，使用已标记的短信数据集进行监督学习，让模型学会区分垃圾短信和正常短信。

模型评估：使用另外一部分测试数据进行模型评估，评估模型的分类准确率、召回率等指标，调整模型参数以提高分类效果。

模型应用：将训练好的模型应用到实际的短信分类任务中，对接收到的短信进行分类，将垃圾短信过滤出去或者放入垃圾箱。


```python
import pandas as pd
```

```python
df = pd.read_excel('垃圾短信数据集.xlsx')
df.head()
```

```python
data = df['短信内容'].tolist()
label = df['是否是垃圾短信'].tolist()
```

```python
print(len(data), len(label))
print(data[0:5])
print(label[0:5])
from collections import Counter
c = Counter(label)
print(c)
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.sans-serif'] = 'SimHei'
matplotlib.rcParams['axes.unicode_minus'] = False
#x = list(c.keys())
x = ['正常短信', '垃圾']
y = list(c.values())
plt.title('垃圾短信类别分布柱形图')
plt.xlabel('类别')
plt.ylabel('数量')
plt.bar(x, y)
plt.show()
```
