## 分词

中文文本处理中要进行分词的主要原因是中文的书写方式不像英文那样使用空格来明确单词的边界。在中文中，词与词之间没有明显的分隔符号，因此需要通过分词来将连续的字序列划分成有意义的词语，以便进行后续的语言分析和处理。分词有助于提高对文本的理解和处理效率，是中文NLP中的基础步骤之一。

另外，中文的词汇构造较为复杂，一个词可能由一个或多个字组成，而且同一个字可以出现在多个词中，这增加了分词的难度。因此，分词在中文NLP中具有重要的作用，对于许多任务，如文本分类、信息检索、机器翻译等，都是必不可少的预处理步骤。

## 停用词

去除停用词是文本处理中的一项常见操作，其主要目的是提高文本处理的效率和质量。停用词是指在文本中频繁出现但通常对文本含义没有太大贡献的词语，例如“的”、“了”、“是”等。去除停用词可以减少文本数据的噪声，使得文本特征更集中于那些具有实际意义的词语，从而有助于提高文本处理任务的准确性和效率。

具体来说，去除停用词有以下几个好处：

减少数据噪声：停用词通常是文本中出现频率最高的词语，但它们往往对文本的实际含义贡献较少。去除停用词可以减少这些对文本分析无用的信息。

减少计算量：去除停用词可以减少文本中需要处理的词语数量，从而降低计算复杂度，提高处理速度。

提高特征表示的质量：去除停用词可以使得文本特征更加集中于那些具有实际意义的词语，有助于提高文本表示的质量。

```python
import jieba #结巴分词
```

```python
#创建停用词列表
def stopwordslist():
    stopwords = [line.strip() for line in open('ChineseStopWords.txt',encoding='UTF-8').readlines()]
    return stopwords
```

```python
# 对句子进行中文分词
def seg_depart(sentence):
    # 对文档中的每一行进行中文分词
    #print("正在分词")
    sentence_depart = jieba.cut(sentence.strip())
    # 创建一个停用词列表
    stopwords = stopwordslist()
    # 输出结果为outstr
    outstr = ''
    # 去停用词
    for word in sentence_depart:
        if word not in stopwords:
            if word != '\t':
                outstr += word
                outstr += " "
    return outstr
```

```python
sentence = "新浪网SINA.COM新闻中心是新浪网最重要的频道之一,24小时滚动报道国内、国际及社会新闻。"
print(seg_depart(sentence))
```

```python
import re

#通过正则表达式只保留中文，可以进一步减少文本中需要处理的词语数量，从而降低计算复杂度，提高处理速度。
def clean_zh_text(text):
    comp = re.compile('[^\u4e00-\u9fa5]')
    return comp.sub('', text)
```

```python
print(seg_depart(clean_zh_text(sentence)))
```

```python
import pandas as pd
```

```python
#读取数据
df = pd.read_excel('垃圾短信数据集.xlsx')
df = df.dropna()
df.head()
```

```python
data = df['短信内容'].tolist()
label = df['是否是垃圾短信'].tolist()
print(len(data), len(label))
```

```python
#分词和去除停用词
data_seg = []
for each in data:
    each_seg = seg_depart(clean_zh_text(each))
    data_seg.append(each_seg)
```

```python
print(data_seg[0:5])
```

```python
import matplotlib
import matplotlib.pyplot as plt
from wordcloud import WordCloud
matplotlib.rcParams['font.sans-serif'] = 'SimHei'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
#绘制垃圾短信的词云图
negative = df[df['是否是垃圾短信'] == 1] #将数据根据label拆分
n_content =  ' '.join(negative['短信内容']) #合并评论内容
n_cut_word = seg_depart(clean_zh_text(n_content))
n_wc = WordCloud(width=1000, height=800,font_path=r'C:\Windows\Fonts\simhei.ttf',max_words=100).generate(n_cut_word)
plt.imshow(n_wc)
plt.title("垃圾短信")
plt.axis('off')
```

```python
#绘制正常短信的词云图
positive = df[df['是否是垃圾短信'] == 0] #将数据根据label拆分
p_content =  ' '.join(positive['短信内容']) #合并评论内容
p_cut_word = seg_depart(clean_zh_text(p_content))
p_wc = WordCloud(width=1000, height=800,font_path=r'C:\Windows\Fonts\simhei.ttf',max_words=100).generate(p_cut_word)
plt.imshow(p_wc)
plt.title("正常短信")
plt.axis('off')
```

```python
from nltk import FreqDist
# 设置图表的大小
plt.figure(figsize=(12, 6))  # 您可以根据需要调整这里的尺寸
# 设置所有文本的默认字体大小
plt.rcParams.update({'font.size': 14})
fdist=FreqDist(n_cut_word.split())
fdist.plot(30, percents=True, title='垃圾短信词语分布百分比图') # 频率分布图
```

```python
#文本长度百分比统计
sentence_length = [len(text) for text in data_seg] #得到数据所有的文本长度
plt.hist(sentence_length, 50, density = True)
plt.xlim(0,max(sentence_length))
plt.xlabel('文本长度')
plt.ylabel('文本数量占比')
plt.show()
```

```python
#保存数据
save_data = {
    '文本': data_seg,
    '标签': label
}
df = pd.DataFrame(save_data)
df.to_csv("分词后data.csv", index=False, encoding="utf-8-sig")
```
